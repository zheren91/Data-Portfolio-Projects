```python
country_url = 'https://companiesmarketcap.com/all-countries/'
page = requests.get(country_url)
soup = BeautifulSoup(page.content)
```

```python
table = soup.find('table', class_= 'default-table table marketcap-table dataTable')
```

```python
link = table.find_all('a',href=True)
# test
link[0]['href']
```

```python

country_urls = ['https://companiesmarketcap.com' + link['href'] for link in table.find_all('a', href = True)]
```

```python
# get all country data
all_countries_data = []
for country_url in country_urls:
    page = requests.get(country_url)
    soup = BeautifulSoup(page.content)
    tickers = [div.get_text() for div in soup.find_all('div',class_='company-code')]
    for ticker in tickers[:10]: # only do the first 10
        try:
            #print(ticker)
            yf_ticker = yf.Ticker(ticker)
            info = yf_ticker.get_info()
            data = [info[key] for key in keys_to_use]
            all_countries_data.append(data)
            sleep(0.5)
        except Exception:
            print(f'skipped {ticker}....')
```

```python
df = pd.DataFrame(all_countries_data, columns = keys_to_use)
```
